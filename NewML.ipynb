```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

print("pandas version:", pd.__version__)
```

```python
from google.colab import files

print("upload games.csv")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
print("Loaded file:", filename)

df_raw = pd.read_csv(filename, low_memory=False)

print("raw shape:", df_raw.shape)
print(df_raw.columns.tolist())

df_raw.head(3)
```

```python
#Fix cols

orig_cols = df_raw.columns.tolist()

# new dataFrame
df_fixed = pd.DataFrame(index=df_raw.index)


df_fixed['AppID'] = df_raw.index

for i in range(1, len(orig_cols)):
    new_col = orig_cols[i]
    old_col = orig_cols[i - 1]
    df_fixed[new_col] = df_raw[old_col]

df_clean = df_fixed.reset_index(drop=True)

print(df_clean[['AppID', 'Name', 'Release date', 'Estimated owners']].iloc[0])
```

```python

# make working copy
df = df_clean.copy()

print(df.shape)
df.head(10)
```

```python
# parse Estimated owners


import numpy as np

def parse_estimated_owners(s):
    """
    mid value
    """
    if pd.isna(s):
        return np.nan

    s = str(s).strip()

    # If '0 - 0'
    if s == '0 - 0':
        return 0

    # Normal
    if '-' in s:
        left, right = s.split('-')
        left = left.strip().replace(',', '')
        right = right.strip().replace(',', '')
        try:
            left = int(left)
            right = int(right)
            return (left + right) / 2
        except ValueError:
            return np.nan

    return np.nan

df['owners_mid'] = df['Estimated owners'].apply(parse_estimated_owners)

df[['AppID', 'Name', 'Estimated owners', 'owners_mid']].head(10)
```

```python
# basic type cleaning


# Ensure AppID is number
df['AppID'] = pd.to_numeric(df['AppID'], errors='coerce').astype('Int64')

# Make sure price is num
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# date must be datetime
df['Release date'] = pd.to_datetime(df['Release date'], errors='coerce')

# take year
df['release_year'] = df['Release date'].dt.year

df[['AppID', 'Name', 'Price', 'Release date', 'release_year', 'owners_mid']].head(10)
```

```python
# reviews features
# Turn to num
df['Positive'] = pd.to_numeric(df['Positive'], errors='coerce')
df['Negative'] = pd.to_numeric(df['Negative'], errors='coerce')

# total reviews
df['total_reviews'] = df['Positive'] + df['Negative']

# If total_reviews is 0, avoid divide 0
df['pos_ratio'] = df['Positive'] / df['total_reviews'].replace(0, np.nan)

df[['Name', 'Positive', 'Negative', 'total_reviews', 'pos_ratio']].head(10)
```

```python
# build a cleaner subset for ML

# important cols
cols_needed = [
    'AppID',
    'Name',
    'owners_mid',
    'Price',
    'release_year',
    'total_reviews',
    'pos_ratio'
]

df_model = df[cols_needed].copy()

print("Before dropping NaN:", df_model.shape)

# dropping NaN
df_model = df_model.dropna(subset=['owners_mid', 'Price', 'release_year', 'pos_ratio'])

print("After dropping NaN:", df_model.shape)

df_model.head(10)
```

```python
#Popularity by percentile ranking

import numpy as np

# delete the old columns to avoid confusion
#for col in ['popularity', 'owners_rank', 'owners_pct']:
#    if col in df_model.columns:
#        df_model.drop(columns=[col], inplace=True)

# 1. Order by owners_mid
df_model = df_model.sort_values('owners_mid').reset_index(drop=True)

N = len(df_model)
print("Number of games in df_model:", N)

# the smallest owners_mid   1/N
# Biggest owners_mid   N/N = 1
df_model['owners_pct'] = (np.arange(N) + 1) / N

# back 25%, low
# front 25%, high
# mid 50%,medium
def label_popularity_pct(p):
    if p <= 0.25:
        return 'low'
    elif p >= 0.75:
        return 'high'
    else:
        return 'medium'

df_model['popularity'] = df_model['owners_pct'].apply(label_popularity_pct)

print("\nPopularity counts:")
print(df_model['popularity'].value_counts())

print("\nPopularity proportion:")
print(df_model['popularity'].value_counts(normalize=True))

df_model[['AppID', 'Name', 'owners_mid', 'owners_pct', 'popularity']].head(10)


```

```python
#Save cleaned dataset
cols_to_save = [
    'AppID',
    'Name',
    'owners_mid',
    'Price',
    'release_year',
    'total_reviews',
    'pos_ratio',
    'owners_pct',
    'popularity'
]

clean_filename = "steam_games_clean_popularity.csv"

df_model[cols_to_save].to_csv(clean_filename, index=False)

print("Saved cleaned data to:", clean_filename)
```
